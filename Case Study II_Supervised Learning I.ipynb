```python
def preprocess(df_horse):
    binary_cols = ["surgery", "surgical_lesion", "cp_data", ]
    df_horse[binary_cols] = df_horse[binary_cols].replace({"yes":1,"no":0})
    df_horse["age"] = df_horse["age"].replace({"young":0, "adult":1})
    df_horse["capillary_refill_time"] = df_horse["capillary_refill_time"].replace({1:0, 2:1})
    df_horse["outcome"] = df_horse["outcome"].replace({"died":0, "euthanized":1, "lived": 2})
```

```python
Y = df_horse["outcome"]
X = df_horse.drop(["outcome"], axis=1)
```

```python
 X= pd.get_dummies(X, dummy_na=True)
```

```python
X.info()
```

```python
Y.head()
```

```python

X = X.apply(lambda x: x.fillna(x.value_counts().index[0]))
```

```python
from sklearn import metrics
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.33, random_state=42)
```

```python
dec_tree = DecisionTreeClassifier()
dec_tree.fit(X_train, y_train)
predicted_outcome = dec_tree.predict(X_test)
metrics.accuracy_score(predicted_outcome, y_test)
```

```python
random_forest = RandomForestClassifier()
random_forest.fit(X_train, y_train)
predicted_outcome = random_forest.predict(X_test)
metrics.accuracy_score(predicted_outcome, y_test)
```
